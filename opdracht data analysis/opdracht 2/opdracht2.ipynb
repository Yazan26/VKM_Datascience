```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

# ik heb een dataset gegenereerd via chatgpt, ik wil die nu importeren en analyseren

```python
#from sklearn.datasets import load_iris
# Load the Iris dataset
data = pd.read_csv('student_data.csv')
# Convert to DataFrame
```

```python
## check na first 
data.isna().sum()
# we zien dat er geen na's in zitten
```

# in deze opdrachten gebruiken wij 3 technieken

- PCA
- T-SNE
- UMAP

# PCA

```python
# split data and target
x = data[['study_hours_per_week','stress_level_1to10','attendance_rate_percent','sleep_hours_per_night','part_time_job_hours_per_week']]
y = data['gpa_1to10']
```

```python
# scale the data
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# apply PCA
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)
```

```python
# visualizing the PCA result
plt.figure(figsize=(6,4))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=y, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.colorbar()
plt.show()
```

# T-SNE

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, random_state=42, perplexity=30)

x_tsne = tsne.fit_transform(x_scaled)
```

```python
# visualizing the PCA result
plt.figure(figsize=(6,4))
plt.scatter(x_tsne[:, 0], x_tsne[:, 1], c=y, cmap='plasma')
plt.xlabel('TSNE Component 1')
plt.ylabel('TSNE Component 2')
plt.title('t-SNE of Iris Dataset')
plt.colorbar()
plt.show()
```
