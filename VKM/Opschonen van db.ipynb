```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

vkm_OG = pd.read_csv(r'C:\Users\Gebruiker\Documents\Code\Datasets\Uitgebreide_VKM_dataset.csv') 

vkm_OG.head()
```

```python
vkm_OG.info()
```

```python
vkm = vkm_OG.drop (columns=['Rood', 'Groen', 'Blauw', 'Geel'])

vkm.head(2)
```

```python
#Nu ga ik de content van de kolommen  description en content volledig lezen, zodat ik weet of ze duplicates van elkaar zijn en dus onnodig.
pd.set_option('display.max_colwidth', None)
vkm.head(2)
```

```python
#Aangezien de content kolom een duplicate is van de description kolom, ga ik deze verwijderen.
vkm = vkm.drop (columns=['content'])
vkm.head(2)
```

```python
vkm['shortdescription'] = vkm['shortdescription'].fillna('Onbekend')
vkm['learningoutcomes'] = vkm['learningoutcomes'].fillna('Onbekend')
```

```python
# check if all datatypes are correct
vkm.dtypes
```

```python
# Change datatype of 'start_date' to datetime
vkm['start_date'] = pd.to_datetime(vkm['start_date'])
vkm.dtypes
```

```python
vkm.head()
```

```python
#nu ga ik alle tekst lowercase maken, zodat er geen duplicates ontstaan door hoofdletters.
vkm['name'] = vkm['name'].str.lower()
vkm['location'] = vkm['location'].str.lower()
vkm['module_tags'] = vkm['module_tags'].str.lower()
vkm.duplicated().sum()
```
