```python
import pandas as pd
df = pd.read_csv("Uitgebreide_VKM_Dataset.csv")
```

```python
df_cleaned = df.drop(columns=["content","Rood","Groen","Blauw","Geel"])
df_cleaned
```

```python
df_cleaned.head()

# now i want to see the dataframe rows with missing values
missing = df_cleaned.isna().sum().sort_values(ascending=False)
missing_percent = (missing / len(df_cleaned) * 100).round(2)
pd.concat([missing, missing_percent], axis=1, keys=["missing_count", "missing_pct"])
# okay i will fill 
# shortdescription missing values with "onbekend"
#learningoutcomes with "onbekend"
df_cleaned['shortdescription'].fillna('onbekend')
df_cleaned['learningoutcomes'].fillna('onbekend')
```

```python
# now i want to convert start_date from string to datetime
df_cleaned['start_date'] = pd.to_datetime(df_cleaned['start_date'], errors='coerce')
```
